```python
import pandas as pd

df = pd.read_csv("../data/raw/SampleSuperstore.csv")

df.head()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Check duplicate rows
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
# Remove leading/trailing spaces

df["Ship Mode"] = df["Ship Mode"].str.strip()
df["Segment"] = df["Segment"].str.strip()
df["Country"] = df["Country"].str.strip()
df["City"] = df["City"].str.strip()
df["State"] = df["State"].str.strip()
df["Region"] = df["Region"].str.strip()
df["Category"] = df["Category"].str.strip()
df["Sub-Category"] = df["Sub-Category"].str.strip()
```

```python
# check for negative values
print("Minimum Sales:", df["Sales"].min())
print("Minimum Quantity:", df["Quantity"].min())
print("Minimum Discount:", df["Discount"].min())
print("Minimum Profit:", df["Profit"].min())
```

```python
# Save cleaned dataset

df.to_csv("../data/processed/cleaned_superstore.csv", index=False)

print("✅ Cleaned dataset saved successfully!")
```
