# Retail Sales Analytics - Exploratory Data Analysis (EDA)

## Objective

The objective of this notebook is to analyze the retail sales dataset and discover meaningful business insights through exploratory data analysis (EDA).

### Tools Used
- Python
- Pandas
- Matplotlib

### Dataset
Sample Superstore Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Display plots inside notebook
%matplotlib inline
```

```python
df = pd.read_csv("../data/processed/cleaned_superstore.csv")

df.head()
```

```python
print("Dataset Shape :", df.shape)

print("\nTotal Sales :", round(df["Sales"].sum(),2))

print("Total Profit :", round(df["Profit"].sum(),2))

print("Average Sales :", round(df["Sales"].mean(),2))

print("Average Profit :", round(df["Profit"].mean(),2))
```

# Analysis 1: Sales by Category

### Business Question

Which product category generates the highest total sales?

```python
sales_category = df.groupby("Category")["Sales"].sum()

sales_category
```

```python
sales_category.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

## Business Insight

The category with the highest total sales contributes the most revenue to the business. This indicates that the company should continue investing in this category through inventory planning and marketing strategies.
