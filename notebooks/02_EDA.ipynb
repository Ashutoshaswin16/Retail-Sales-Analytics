# Retail Sales Analytics - Exploratory Data Analysis (EDA)

## Objective

The objective of this notebook is to analyze the retail sales dataset and discover meaningful business insights.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../data/processed/cleaned_superstore.csv")

df.head()
```

```python
print("Dataset Shape :", df.shape)

print("\nTotal Sales :", round(df["Sales"].sum(),2))

print("Total Profit :", round(df["Profit"].sum(),2))

print("Average Sales :", round(df["Sales"].mean(),2))

print("Average Profit :", round(df["Profit"].mean(),2))
```

# Analysis 1: Sales by Category

### Business Question

Which product category generates the highest sales?

```python
sales_category = df.groupby("Category")["Sales"].sum()

print(sales_category)
```

```python
sales_category.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

## Business Insight

Technology generated the highest sales among all categories.

This indicates that technology products contribute significantly to the company's revenue.

```python
print(sales_category)
```

## Business Insight

- Technology generated the highest total sales (836,154.03), making it the company's top revenue-generating category.
- Furniture ranked second in sales, while Office Supplies generated the lowest revenue.
- The business can focus on expanding its Technology product line and marketing efforts to maximize revenue.

# Analysis 2: Profit by Region

### Business Question

Which region generates the highest total profit?

```python
profit_region = df.groupby("Region")["Profit"].sum()

print(profit_region)
```

```python
profit_region.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")

plt.show()
```

```python
profit_region = df.groupby("Region")["Profit"].sum()

print(profit_region)
```

```python
profit_region.plot(
    kind="bar",
    figsize=(8,5),
)

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.show()
```

## Business Insight

- The West region generated the highest total profit (108,418.45), making it the most profitable region.
- The East region ranked second with a profit of 91,522.78.
- The Central region recorded the lowest profit, indicating potential opportunities for improving pricing, sales strategies, or operational efficiency in that region.

# Analysis 3: Sales by Customer Segment

### Business Question

Which customer segment contributes the highest sales?

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

print(segment_sales)
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

print(segment_sales)
```

```python
segment_sales.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6),
    startangle=90
)

plt.title("Sales by Customer Segment")
plt.ylabel("")

plt.show()
```

## Business Insight

- The Consumer segment generated the highest total sales (1,161,401), making it the company's primary revenue source.
- The Corporate segment ranked second, contributing a significant share of total sales.
- The Home Office segment generated the lowest sales, suggesting an opportunity to improve marketing and customer engagement for this segment.

# Analysis 4: Ship Mode Distribution

### Business Question

Which shipping mode is used most frequently by customers?

```python
ship_mode = df["Ship Mode"].value_counts()

print(ship_mode)
```

```python
ship_mode.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Number of Orders by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")

plt.show()
```

## Key Findings

**Observation:**
The Consumer segment generated the highest total sales.

**Business Impact:**
This indicates that individual consumers are the company's primary revenue source.

**Recommendation:**
The company should continue investing in products, promotions, and loyalty programs targeted at consumer customers while exploring strategies to grow the Corporate and Home Office segments.

```python
ship_mode = df["Ship Mode"].value_counts()

print(ship_mode)
```

```python
ship_mode.plot(
    kind="bar",
    figsize=(8,5),
    color="skyblue"
)

plt.title("Number of Orders by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")

plt.xticks(rotation=0)

plt.show()
```

## Key Findings

### Observation
- Standard Class is the most frequently used shipping mode with 5,968 orders.
- Same Day shipping is the least frequently used option with only 543 orders.

### Business Impact
- Most customers prefer the lower-cost Standard Class shipping option.
- Premium shipping options are used less frequently, indicating lower demand or higher costs.

### Recommendation
- Continue optimizing Standard Class delivery since it handles most orders.
- Consider promotional offers or pricing strategies to encourage greater adoption of faster shipping options.

# Analysis 5: Top 10 States by Sales

### Business Question

Which states generate the highest total sales?

```python
top_states = (
    df.groupby("State")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_states)
```

```python
plt.figure(figsize=(10,6))

top_states.plot(
    kind="barh",
    color="steelblue"
)

plt.title("Top 10 States by Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("State")

plt.gca().invert_yaxis()

plt.show()
```

## Key Findings

### Observation

- California generated the highest total sales (457,687.63), followed by New York (310,876.27).
- Texas ranked third with total sales of 170,188.05.
- The top three states contribute a significant portion of overall company revenue.

### Business Impact

- California and New York are the strongest revenue-generating markets.
- These states should remain a priority for inventory planning, marketing campaigns, and customer retention initiatives.

### Recommendation

- Continue investing in high-performing states while identifying strategies to improve sales in lower-performing states.

# Analysis 7: Correlation Between Numerical Features

## Business Question

How are Sales, Quantity, Discount, and Profit related?

```python
correlation = df[["Sales", "Quantity", "Discount", "Profit"]].corr()

print(correlation)
```

```python
plt.figure(figsize=(6,5))

plt.imshow(correlation, cmap="coolwarm", interpolation="nearest")

plt.colorbar()

plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=45)
plt.yticks(range(len(correlation.columns)), correlation.columns)

plt.title("Correlation Matrix")

plt.show()
```

## Key Findings

### Observation
The correlation matrix shows how strongly the numerical variables are related.

### Business Impact
Understanding these relationships helps identify factors that influence sales and profit.

### Recommendation
Further analysis should focus on variables that have stronger positive or negative correlations with profit.

# Analysis 9: Top 10 Sub-Categories by Profit

Revenue alone does not determine business success. Profitability should also be considered when making inventory and pricing decisions.

### Recommendation

Increase focus on highly profitable sub-categories such as Copiers and Phones while investigating why high-selling products like Chairs generate comparatively lower profit.

# Analysis 10: Loss-Making Sub-Categories

## Business Question

Which sub-categories generate the lowest profit or losses?

```python
loss_subcategories = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .sort_values()
      .head(10)
)

print(loss_subcategories)
```

```python
plt.figure(figsize=(10,6))

loss_subcategories.plot(
    kind="barh",
    color="red"
)

plt.title("Bottom 10 Sub-Categories by Profit")
plt.xlabel("Profit")
plt.ylabel("Sub-Category")

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()

plt.show()
```

```python
loss_subcategories = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .sort_values()
      .head(10)
)

print(loss_subcategories)
```

```python
plt.figure(figsize=(10,6))

loss_subcategories.plot(
    kind="barh",
    color="crimson"
)

plt.title("Bottom 10 Sub-Categories by Profit", fontsize=14)
plt.xlabel("Total Profit", fontsize=12)
plt.ylabel("Sub-Category", fontsize=12)

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()

plt.show()
```

## Key Findings

### Observation

- Tables recorded the largest overall loss (-17,725.48).

- Several sub-categories generated only a small amount of profit despite contributing to sales.

### Business Impact

Loss-making products reduce the company's overall profitability and may indicate issues with pricing, discount strategies, or operating costs.

### Recommendation

The company should investigate why Tables, Bookcases, and Supplies are unprofitable. Reviewing supplier costs, pricing strategies, and discount policies could improve profitability.
